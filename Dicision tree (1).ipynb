```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
clf=DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=20)
```

```python
clf=clf.fit(X_train,y_train)
```

```python
from sklearn import tree
```

```python
classnames=['Retain','not-retain']
```

```python
fig=plt.figure(figsize=(60,60))
tree.plot_tree(clf,feature_names=X_train.columns,class_names=classnames,filled=True);
```

```python
from sklearn.metrics import confusion_matrix
train_tree_predict=clf.predict(X_train)
```

```python
classname=['pass','fail']
confusionmatrix=confusion_matrix(y_train,train_tree_predict)
sns.heatmap(confusionmatrix,annot=True,yticklabels=classname,xticklabels=classname,fmt='g',cmap='Greens')
```

```python
#evaluating roc
metrics.roc_auc_score(y_train,train_tree_predict)
```

```python
#for test data
from sklearn import metrics
from sklearn.metrics import confusion_matrix
test_tree_predict=clf.predict(X_test)
```

```python
classnames=['Pass','Fail']
confusionmatrix=confusion_matrix(y_test,test_tree_predict)
sns.heatmap(confusionmatrix,annot=True,yticklabels=classname,xticklabels=classname,fmt='g',cmap='Greens')
```

```python
metrics.roc_auc_score(y_test,test_tree_predict)
```
