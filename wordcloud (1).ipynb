```python
import os
import pathlib
import nltk
import numpy as np
import seaborn as sn
import pandas as pd
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
import urllib.request
from wordcloud import WordCloud  # Corrected import statement

nltk.download('punkt')
```

```python
textfile=open('C:/Users/Priyanka Hegde/Downloads/pdfcoffee.com_crooked-house-by-agatha-christie-pdf-free.txt','r').read()
print(textfile)
```

```python
words=word_tokenize(textfile)
print(words)
```

```python
print(f'total number of words in the file={len(words)}')
```

```python
f=FreqDist(words)
f.most_common(10)
```

```python
word_no_punc=[]
for word in words:
    if word.isalpha():
        word_no_punc.append(word.lower())
print(f"total number of words without punctuation is{len(word_no_punc)}")
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words=stopwords.words("english")
print(stop_words)
```

```python
clean_word=[]
for word in word_no_punc:
    if word not in stop_words:
        clean_word.append(word)
print(f"Clean word without stopwords={len(clean_word)}")
print(clean_word)
```

```python
f=FreqDist(clean_word)
f.most_common(10)
```

```python
clean_words=" ".join(clean_word)
wc=WordCloud(width=800, height=500, background_color='white', max_words=100)
wc.generate(clean_words)
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
def transform_format(val) :
    if val==0 :
        return 255
    else:
        return val
```

```python
transformed_mask=np.ndarray((mask.shape[0],mask.shape[1]),np.int32)
for i in range(len(mask)):
    transformed_mask[i]=list(map(transform_format,mask[i]))
print(transformed_mask)
```
